```python
import os
import cv2
from google.colab import drive

# 구글 드라이브 마운트
drive.mount('/content/drive')

import os
os.chdir("/content/drive/")
!ls

# 동영상 파일들이 들어있는 폴더 경로
folder_path = "/content/drive/MyDrive/video"

# 이미지 저장 폴더 경로
image_folder = "/content/drive/MyDrive/imagedata"

# 폴더 내의 모든 파일을 가져옴
file_list = os.listdir(folder_path)

# 이미지 저장 폴더 생성
os.makedirs(image_folder, exist_ok=True)

for file_name in file_list:
    # 파일 확장자 확인
    if file_name.endswith(".mp4") or file_name.endswith(".avi"):
        # 영상 파일 경로
        video_path = os.path.join(folder_path, file_name)

        # 영상 처리를 위한 초기 설정
        video_capture = cv2.VideoCapture(video_path)
        fps = video_capture.get(cv2.CAP_PROP_FPS)
        frame_count = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))
        interval = 5 * fps  # 5초 간격

        # 이미지 추출 및 저장
        count = 0
        success, image = video_capture.read()
        while success:
            if count % interval == 0:
                image_path = os.path.join(image_folder, f"{file_name}_frame_{count}.jpg")
                cv2.imwrite(image_path, image)
            success, image = video_capture.read()
            count += 1

        # 영상 캡처 객체 해제
        video_capture.release()
```
